```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
```

```python
df.drop('deck',axis=1, inplace=True)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
sns.scatterplot(x='age',y='fare',data=df1)
```

```python
sns.scatterplot(x='age',y='fare',data=df1,hue='sex')
```

```python
sns.histplot(df1,x='fare',hue='pclass',bins=10)
```

```python
sns.histplot(df1,x='fare',hue='pclass',bins=10,kde=True)
```

```python
sns.displot(df1['fare'],aspect=3)
```

```python
sns.displot(df1,x='age',hue='sex',aspect=3)
```

```python
sns.displot(df1,x='age',hue='sex',col='pclass',aspect=1)
```
